## Step 1 : 📥 Load Data

```python
# Prepare Environment

import numpy as np      # Numerical computing library (arrays, math functions)
import pandas as pd     # Data analysis and manipulation library
import seaborn as sns   # Statistical data visualization library
import matplotlib.pyplot as plt  # Main plotting library for charts and graphs
```

```python
# 1st step : Read Data -> read_csv

Data = pd.read_csv(r"D:\Courses\DEPI R4 - Microsoft ML\Graduation Project\Phase 0 - About Dataset\Dataset\Metro.csv")
```

```python
# Searching for duplicate values

Duplicates = Data.duplicated().sum()

print("Number of Duplicates:", Duplicates)
```

```python
# Searching for null values

Nulls = Data.isnull().sum()

Nulls_DF = pd.DataFrame(Nulls, columns=["Number of Nulls"])

Nulls_DF.index.name = "Feature"

display(Nulls_DF)
```

## Step 3 : 🧹 Data Cleaning

```python
# Drop not important columns

if "Unnamed: 0" in Data.columns:
    Data.drop(columns=['Unnamed: 0'], inplace=True)
    print("Done.")
```

```python
# Standardize column names

Data.columns = [col.lower().replace(' ', '_').strip() for col in Data.columns]
```

```python
# Convert timestamp column to datetime format for proper time-based analysis

Data['timestamp'] = pd.to_datetime(Data['timestamp'])
```

```python
# Resampling Time-Series Data to 10-Second Intervals

Data = Data.set_index('timestamp')

Data.resample('10s').mean()
```

```python
# Print the list of columns after cleaning

print("Columns after cleaning:", Data.columns.tolist())
```

## Step 4 : 🪙 Feature Engineering

```python
Data['hour'] = Data.index.hour

Data['day_of_week'] = Data.index.dayofweek

Data['pressure_delta'] = Data['tp3'] - Data['tp2']

Data['power_indicator'] = Data['motor_current'] * Data['reservoirs']
```

```python
Data['oil_temp_rolling'] = Data['oil_temperature'].rolling(window=60).mean()

Data['pressure_change'] = Data['tp3'].diff()

Data['current_per_pressure'] = Data['motor_current'] / (Data['tp2'] + 0.1)

Data.dropna(inplace=True)
```

## Step 5 : 🔎 Data Analysis

```python
Gap_Series = Data.index.to_series().diff().dt.total_seconds()

Gap_Series = Gap_Series.fillna(0)

plt.figure(figsize=(12, 6))

sns.histplot(Gap_Series, bins=50, kde=False, color='teal', log_scale=(False, True))

plt.title('Distribution of Time Gaps (Log Scale)', fontsize=15)
plt.xlabel('Gap Duration (Seconds)', fontsize=12)
plt.ylabel('Frequency (Logarithmic)', fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
Top_Gaps = Gap_Series.sort_values(ascending=False).head(5)

print("--- Top 5 Data Gaps (In Seconds) ---")
print(Top_Gaps)
```

```python
Corr_Matrix = Data.corr()

plt.figure(figsize=(16, 12))
sns.heatmap(Corr_Matrix, annot=True, fmt=".2f", cmap='RdBu_r', center=0, linewidths=0.5)
plt.title("Correlation Heatmap: How Sensors Relate to Each Other", fontsize=16)
plt.show()
```

```python
important_sensors = ['tp2', 'tp3', 'h1', 'oil_temperature', 'motor_current']

corr_matrix = Data[important_sensors].corr()

import seaborn as sns
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='RdBu_r', center=0)
plt.title('Mechanical Sensors Correlation Map')
plt.show()
```

```python
df_sample = Data.tail(1000) 

cols_to_plot = ['tp2', 'tp3', 'h1', 'oil_temperature', 'motor_current']
plt.figure(figsize=(18, 12))

for i, col in enumerate(cols_to_plot):
    ax = plt.subplot(len(cols_to_plot), 1, i+1)
    
    ax.plot(df_sample.index, df_sample[col], color='steelblue', linewidth=0.8, label='Raw Signal')
    
    ax.plot(df_sample.index, df_sample[col].rolling(window=20).mean(), 
            color='orange', linewidth=1.5, label='Trend (SMA 20)')
    
    ax.set_title(f'Sensor {col} Reading over Time', loc='left', fontsize=12)
    ax.legend(loc='upper right', fontsize=8)
    ax.grid(True, alpha=0.3)
    
    if i < len(cols_to_plot) - 1:
        ax.set_xticklabels([])

plt.xlabel('Time')
plt.suptitle('Compressor Sensors Behavior Analysis', fontsize=16, y=0.98)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
```
